```python
# Detect outliers in the Units_Sold column using the IQR method. What are the lower and upper bounds?
q1=df["Units_Sold"].quantile(0.25)
q1
```

```python
q3=df["Units_Sold"].quantile(0.75)
q3
```

```python
IQR=q3-q1
lower_boundary=q1-1.5*IQR
lower_boundary
```

```python
upper_boundary=q3+1.5*IQR
upper_boundary
```

```python
#Remove rows containing outliers in the Units_Sold column and compare the original and cleaned dataset sizes
df2=df[(df["Units_Sold"]>lower_boundary)&(df["Units_Sold"]<upper_boundary)]
df2
```

```python
#Visualize the outliers using a box plot before and after treatment
sns.boxplot(data=df)
```

```python
sns.boxplot(data=df2)
```
