```python
import pandas as pd
df1=pd.read_csv("/content/departments.csv")
df1
```

```python
df2=pd.read_csv("/content/employees.csv")
df2
```

```python
#Inner Join: Perform an inner join between the employees and departments datasets
#to get information about employees along with their department names and locations.
inner=pd.merge(df2,df1,on="Department_ID",how="inner")
inner
```

```python
#Left Join: Perform a left join between the employees and departments datasets.
left=pd.merge(df2,df1,on="Department_ID",how="left")
left
```

```python
#Right Join: Perform a right join between the departments and employees datasets.
#Why might this join be useful in certain scenarios?
right=pd.merge(df1,df2,on="Department_ID",how="right")
right
```

```python
#Full Outer Join: Perform a full outer join between the employees and departments datasets.
outer=pd.merge(df2,df1,on="Department_ID",how="outer")
outer
```

```python
# Group Employees by Department:
# Group the employees dataset by Department_ID and calculate:
# Total salary in each department.
total_salary=df2.groupby("Department_ID")["Salary"].sum()
total_salary
```

```python
# Average salary in each department.
avg_salary=df2.groupby("Department_ID")["Salary"].mean()
avg_salary
```

```python
#Join the datasets and group by Location to find:
# The number of employees in each location.
outer=pd.merge(df2,df1,on="Department_ID",how="outer")
a=outer.groupby("Location")["Employee_ID"].count()
a
```

```python
# Total salary of employees in each location.
b=outer.groupby("Location")["Salary"].sum()
b
```

```python
#Filter employees with a salary above $50,000 and group them by
#Department_ID. Calculate the average salary for these employees in each department
filter=df2["Salary"]>50000
b=df2[filter]
c=b.groupby("Department_ID")["Salary"].mean()
c
```
