```python
# Concatenate files into two master datasets: one for weekdays and one for weekends.
weekdays=pd.concat([amsterdam_weekdays,athens_weekdays,barcelona_weekdays,berlin_weekdays,budapest_weekdays,lisbon_weekdays,london_weekdays,paris_weekdays,rome_weekdays,vienna_weekdays],ignore_index=True)
weekdays.head(5)
```

```python
# Combine the datasets into a single unified dataset with an additional column indicating "day_type" as either 'weekday' or 'weekend'.
weekends["data_type"]="weekend"
weekdays["data_type"]="weekdays"
```

```python
combined_data=pd.concat([weekdays,weekends],ignore_index=True)
combined_data.head(5)
```

 columns).
> The categorical columns are  
1. room_type: the type of the accommodation
2. room_shared: dummy variable for shared rooms
3. room_private: dummy variable for private rooms
4. host_is_superhost: dummy variable for superhost status and
 data_type

The numerical columns are
1. realSum: the full price of accommodation for two people and two nights in EUR
2. person_capacity: the maximum number of guests
3. multi: dummy variable if the listing belongs to hosts with 2-4 offers
4. biz: dummy variable if the listing belongs to hosts with more than 4 offers
5. cleanliness_rating: cleanliness rating
6. guest_satisfaction_overall: overall rating of the listing
7.  bedrooms: number of bedrooms (0 for studios)
8.  dist: distance from city centre in km
9.  metro_dist: distance from nearest metro station in km
10. attr_index: attraction index of the listing location
11. attr_index_norm: normalised attraction index (0-100)
12. rest_index: restaurant index of the listing location
13. rest_index_norm: normalised restaurant index (0-100)
14. lng: longitude of the listing location
15. lat: latitude of the listing location
